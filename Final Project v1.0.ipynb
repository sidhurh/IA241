```python
%matplotlib inline
import pandas
```

Data Source(as of May 2): https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide 
        

```python
df = pandas.read_excel('s3://sidhu-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-02.xlsx')
df
```

# The Sum of Deaths Per Day

```python
sum_death_day = df.groupby('dateRep').sum()['deaths']

sum_death_day.plot()
```

There is a peak of COVID-19 deaths around mid-April.
COVID-19 deaths started to climb around mid-March.

# Top 25 Countries With the Lowest Amount of Deaths

```python
low_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
low_death_by_country.nsmallest(25).plot.bar()
```

The top twenty-five countries with the lowest amount of deaths all have zero recorded deaths due to COVID-19.

# Top 25 Countries With the Highest Amount of Deaths

```python
top_death_by_country = df.groupby('countriesAndTerritories').sum()['deaths']
top_death_by_country.nlargest(25).plot.bar()
```

The top 5 countries with the most deaths have over 200,000 deaths while the rest have below 100,000.

# Top 25 Countries with the Lowest Amount of Cases

```python
low_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
low_cases_by_country.nsmallest(25).plot.bar()
```

Comoros has the least amount of cases with only 1, while Anguila has the second lowest amount of cases with 3.

# Top 25 Countries with the Highest Amount of Cases

```python
top_cases_by_country = df.groupby('countriesAndTerritories').sum()['cases']
top_cases_by_country.nlargest(25).plot.bar()
```

The United States has the most amount of COVID-19 cases with over 1,000,000 cases while the second highest country Spain only has around 200,000 cases. That is an 800,000 case difference between the top two countries with the most cases.

# Conclusion

The United States has the most cases and deaths from COVID-19 in the world, by a lead over other countries of around 800,000 cases and 300,000 deaths. Comoros has the least amount of cases with 1. There are at least 25 countries that have the lowest amount of COVID-19 deaths with 0. COVID-19 deaths have peaked in mid-April and have started declining.
